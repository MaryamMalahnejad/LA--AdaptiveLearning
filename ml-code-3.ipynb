```python
# Install necessary libraries
!pip install plotly dash dash-bootstrap-components

# Import libraries
import plotly.express as px
import dash
from dash import dcc, html
from dash.dependencies import Input, Output

# Sample data
# Assuming df is your DataFrame
# df = ...

# Initialize the Dash app
app = dash.Dash(__name__)

# Define the layout of the app
app.layout = html.Div([
    html.H1("Interactive Dashboard"),

    # Dropdown for selecting age range
    dcc.Dropdown(
        id='age-dropdown',
        options=[
            {'label': 'All Ages', 'value': 'all'},
            {'label': '18-25', 'value': '18-25'},
            {'label': '26-35', 'value': '26-35'},
            {'label': '36-45', 'value': '36-45'},
            {'label': '46+', 'value': '46+'}
        ],
        value='all',
        style={'width': '50%'}
    ),

    # Bar chart for age distribution
    dcc.Graph(id='age-distribution'),

    # Pie chart for gender distribution
    dcc.Graph(id='gender-distribution'),

    # Bar chart for study field distribution
    dcc.Graph(id='study-field-distribution')
])

# Define callback functions to update charts based on dropdown selection
@app.callback(
    Output('age-distribution', 'figure'),
    Output('gender-distribution', 'figure'),
    Output('study-field-distribution', 'figure'),
    Input('age-dropdown', 'value')
)
def update_charts(selected_age):
    filtered_df = df if selected_age == 'all' else df[df['Age'].between(18, 25) if selected_age == '18-25' else
                                                       df['Age'].between(26, 35) if selected_age == '26-35' else
                                                       df['Age'].between(36, 45) if selected_age == '36-45' else
                                                       df['Age'] >= 46]

    # Age Distribution Bar Chart
    age_fig = px.histogram(filtered_df, x='Age', nbins=10, title='Age Distribution')

    # Gender Distribution Pie Chart
    gender_fig = px.pie(filtered_df, names='Gender', title='Gender Distribution')

    # Study Field Distribution Bar Chart
    study_field_fig = px.bar(filtered_df, x='Study_Field', title='Study Field Distribution')

    return age_fig, gender_fig, study_field_fig

# Run the app
if __name__ == '__main__':
    app.run_server()
```
